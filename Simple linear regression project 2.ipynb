```python
import numpy as np
import matplotlib.pyplot as plt

# Data
hours = np.array([2.5, 5.1, 3.2, 8.5, 3.5, 1.5, 9.2, 5.5, 8.3, 2.7, 7.7, 5.9, 4.5, 3.3, 1.1, 8.9, 2.5, 1.9, 6.1, 7.4, 2.7, 4.8, 3.8, 6.9, 7.8])
scores = np.array([21, 47, 27, 75, 30, 20, 88, 60, 81, 25, 85, 62, 41, 42, 17, 95, 30, 24, 67, 69, 30, 54, 35, 76, 86])

# Calculate the slope and intercept of the regression line
X_mean = np.mean(hours)
y_mean = np.mean(scores)

numerator = np.sum((hours - X_mean) * (scores - y_mean))
denominator = np.sum((hours - X_mean) ** 2)

slope = numerator / denominator
intercept = y_mean - slope * X_mean

# Make predictions
predictions = slope * hours + intercept

# Plot the original data points and the regression line
plt.scatter(hours, scores, color='blue', label='Original data')
plt.plot(hours, predictions, color='red', label='Regression line')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()

# Print the coefficients
print('Intercept:', intercept)
print('Slope:', slope)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming you have loaded the dataset into a DataFrame named 'analysis'
# If not, load the dataset using pd.read_csv()

# Extracting the independent variable (feature) and dependent variable (target)
X = analysis['Hours'].values  # Feature: 'Hours'
y = analysis['Scores'].values # Target: 'Scores'

# Calculate the slope and intercept of the regression line
X_mean = np.mean(X)
y_mean = np.mean(y)

numerator = np.sum((X - X_mean) * (y - y_mean))
denominator = np.sum((X - X_mean) ** 2)

slope = numerator / denominator
intercept = y_mean - slope * X_mean

# Make predictions
predictions = slope * X + intercept

# Plot the original data points and the regression line
plt.scatter(X, y, color='blue', label='Original data')
plt.plot(X, predictions, color='red', label='Regression line')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()

# Print the coefficients
print('Intercept:', intercept)
print('Slope:', slope)
```

```python
# Let's categorize scores into three groups: low, medium, and high
bins = [0, 40, 60, 100]
labels = ['Low', 'Medium', 'High']
analysis['Score_Category'] = pd.cut(analysis['Scores'], bins=bins, labels=labels)

# Count the number of scores in each category
score_counts = analysis['Score_Category'].value_counts()

# Plot a pie chart
plt.pie(score_counts, labels=score_counts.index, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightcoral', 'lightskyblue'])
plt.axis('equal')
plt.title('Pie Chart of Score Categories')
plt.show()
```
