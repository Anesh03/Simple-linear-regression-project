```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Data
years_experience = np.array([1.1, 1.3, 1.5, 2.0, 2.2, 2.9, 3.0, 3.2, 3.2, 3.7, 3.9, 4.0, 4.0, 4.1, 4.5, 4.9, 5.1, 5.3, 5.9, 6.0, 6.8, 7.1, 7.9, 8.2, 8.7, 9.0, 9.5, 9.6, 10.3, 10.5])
salary = np.array([39343, 46205, 37731, 43525, 39891, 56642, 60150, 54445, 64445, 57189, 63218, 55794, 56957, 57081, 61111, 67938, 66029, 83088, 81363, 93940, 91738, 98273, 101302, 113812, 109431, 105582, 116969, 112635, 122391, 121872])

# Add a constant term for the intercept
X = sm.add_constant(years_experience)

# Fit the regression model
model = sm.OLS(salary, X).fit()

# Extract the slope and intercept
slope = model.params[1]
intercept = model.params[0]

# Predicted values
predicted_salary = model.predict(X)

# Residuals
residuals = salary - predicted_salary

# Plot 1: Scatter plot with regression line
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(years_experience, salary, label='Data')
plt.plot(years_experience, intercept + slope * years_experience, color='red', label='Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Simple Linear Regression')
plt.legend()

# Plot 2: Residuals plot
plt.subplot(1, 2, 2)
plt.scatter(predicted_salary, residuals, label='Residuals')
plt.axhline(y=0, color='gray', linestyle='--')
plt.xlabel('Predicted Salary')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.legend()

# Show plots
plt.tight_layout()
plt.show()
```
